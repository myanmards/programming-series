There are two steps to get the webdata
* First is to request the page data by using requests library and get method
* Second is to parse the html data using BeautifulSoup library

```python
import requests 
from bs4 import BeautifulSoup

page = requests.get('http://www.gad.gov.mm/en/content/data')
soup = BeautifulSoup(page.content, 'html.parser')
print(soup)
```

```python
print(soup.find('table'))
```

It is possible to look through content of the web data using html tag
* For example, below cell will look through table tag and loop through all the table row using tr tag

```python
table = soup.find('table')
trs = table.find_all('tr')

for tr in trs: # for every table row
    for td in tr.find_all('td'):
        print(td)
```

<b>Note: get_text method is used to get textual information inside each table data</b>

```python
table = soup.find('table')
trs = table.find_all('tr')
rows = []
for tr in trs: # for every table row
    for td in tr.find_all('td'):
        rows.append((td.get_text(strip=True)))
rows
```

```python
table = soup.find('table')
trs = table.find_all('tr')
rows = []
for tr in trs: # for every table row
    rows.append([td.get_text(strip=True) for td in tr.find_all('td')])     
rows
```

<b>Note: With pandas, we can create a dataframe for scraped data as shown below</b>

```python
import pandas as pd
dftable = pd.DataFrame(rows[2:], columns=rows[0])
dftable.head()
```

```python
dftable.to_csv('gad.csv')
```
